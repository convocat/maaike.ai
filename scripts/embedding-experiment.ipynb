# Embedding experiment: comparing three local models on garden content

This notebook compares three embedding models (nomic-embed-text, bge-m3, embeddinggemma) on 10 items from the digital garden. It tests two approaches:

1. **Body text only**: just the markdown content
2. **With metadata**: title + description + tags prepended to the body

**Prerequisites**: Ollama running locally with the three models pulled:
```
ollama pull nomic-embed-text
ollama pull bge-m3
ollama pull embeddinggemma:300m
```
